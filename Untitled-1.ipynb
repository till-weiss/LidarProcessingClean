```python
import numpy as np
import rasterio
import geopandas as gpd
import matplotlib.pyplot as plt

from rasterio.warp import reproject, Resampling
from rasterio.features import rasterize

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# ============================================================
# FILE PATHS
# ============================================================


dem_reference_path = "/isipd/projects/Response/GIS_RS_projects/Masterarbeit_Till_Weiss/results/Inuvik_Airport_2023/DSM/Inuvik_Airport_DSM_2m.tif"

dem_target_path = "/isipd/projects/Response/GIS_RS_projects/Masterarbeit_Till_Weiss/results/Inuvik_Airport_2025/DSM/Inuvik_Airport_DSM_2m.tif"

stable_ground_gpkg = "/isipd/projects/Response/GIS_RS_projects/Masterarbeit_Till_Weiss/AOI/StableReferences/Inuvik_Airport_Stable_Ground.gpkg"

output_path = "./"

output_path = "./"

# ============================================================
# HELPER FUNCTIONS
# ============================================================

def compute_nmad(data):
    return 1.4826 * np.median(
        np.abs(data - np.median(data))
    )


def align_to_reference(src_path, ref_path):

    with rasterio.open(ref_path) as ref:

        ref_data = ref.read(1)

        ref_transform = ref.transform
        ref_crs       = ref.crs
        ref_shape     = ref_data.shape

    with rasterio.open(src_path) as src:

        aligned = np.empty(ref_shape, dtype=np.float32)

        reproject(
            source=rasterio.band(src, 1),
            destination=aligned,
            src_transform=src.transform,
            src_crs=src.crs,
            dst_transform=ref_transform,
            dst_crs=ref_crs,
            resampling=Resampling.bilinear
        )

    return aligned, ref_transform, ref_crs


# ============================================================
# STABLE-GROUND COREGISTRATION
# ============================================================

def stable_ground_coregistration(
    reference_dem,
    target_dem,
    stable_mask
):

    valid = (
        np.isfinite(reference_dem) &
        np.isfinite(target_dem) &
        stable_mask
    )

    dz = target_dem[valid] - reference_dem[valid]

    median_bias = np.median(dz)

    corrected = target_dem - median_bias

    dz_after = corrected[valid] - reference_dem[valid]

    nmad_before = compute_nmad(dz)
    nmad_after  = compute_nmad(dz_after)

    print("\n===================================")
    print("STABLE-GROUND COREGISTRATION")
    print("===================================")
    print(f"Median vertical bias: {median_bias:.3f} m")
    print(f"NMAD before correction: {nmad_before:.3f} m")
    print(f"NMAD after correction:  {nmad_after:.3f} m")

    return corrected, median_bias


# ============================================================
# ALIGN DEMS
# ============================================================

print("\nAligning DEMs...")

reference_dem, transform, crs = align_to_reference(
    dem_reference_path,
    dem_reference_path
)

target_dem, _, _ = align_to_reference(
    dem_target_path,
    dem_reference_path
)

# ============================================================
# REMOVE INVALID VALUES
# ============================================================

reference_dem[reference_dem < -100] = np.nan
target_dem[target_dem < -100]       = np.nan

# ============================================================
# LOAD STABLE GROUND
# ============================================================

print("\nLoading stable-ground mask...")

stable_gdf = gpd.read_file(stable_ground_gpkg).to_crs(crs)

stable_gdf["geometry"] = stable_gdf.buffer(2)

stable_mask = rasterize(
    [(geom, 1) for geom in stable_gdf.geometry],
    out_shape=reference_dem.shape,
    transform=transform,
    fill=0,
    dtype="uint8"
).astype(bool)

# ============================================================
# COREGISTER DEMS
# ============================================================

target_dem_corr, median_bias = stable_ground_coregistration(
    reference_dem,
    target_dem,
    stable_mask
)

# ============================================================
# CREATE RESIDUALS
# ============================================================

residuals = target_dem_corr - reference_dem

# ============================================================
# VALID MASK
# ============================================================

mask = (
    np.isfinite(reference_dem) &
    np.isfinite(target_dem_corr) &
    stable_mask
)

ref = reference_dem[mask]
tar = target_dem_corr[mask]
res = residuals[mask]

# ============================================================
# OUTLIER FILTER
# ============================================================

valid = np.abs(res) < 5

ref = ref[valid]
tar = tar[valid]
res = res[valid]

# ============================================================
# METRICS
# ============================================================

mae    = mean_absolute_error(ref, tar)
rmse   = np.sqrt(mean_squared_error(ref, tar))
r2     = r2_score(ref, tar)

median = np.median(res)
nmad   = compute_nmad(res)

print("\n===================================")
print("FINAL METRICS")
print("===================================")

print(f"MAE: {mae:.3f} m")
print(f"RMSE: {rmse:.3f} m")
print(f"R²: {r2:.3f}")

print(f"Median residual: {median:.3f} m")
print(f"NMAD: {nmad:.3f} m")

# ============================================================
# RANDOM SAMPLE FOR SCATTER
# ============================================================

n = min(20000, len(ref))

idx = np.random.choice(
    len(ref),
    n,
    replace=False
)

# ============================================================
# FIGURE 1 — SCATTER + HISTOGRAM
# ============================================================

fig, axs = plt.subplots(
    2,
    1,
    figsize=(10, 12)
)

# ------------------------------------------------------------
# SCATTER
# ------------------------------------------------------------

axs[0].scatter(
    ref[idx],
    tar[idx],
    s=6,
    alpha=0.3
)

min_val = np.nanpercentile(ref, 1)
max_val = np.nanpercentile(ref, 99)

axs[0].plot(
    [min_val, max_val],
    [min_val, max_val],
    'r--'
)

axs[0].set_xlabel("Reference DEM Elevation (m)")
axs[0].set_ylabel("Corrected DEM Elevation (m)")

axs[0].set_title(
    "Stable-Ground DEM Comparison"
)

axs[0].text(
    0.05,
    0.95,
    (
        f"MAE: {mae:.3f} m\n"
        f"RMSE: {rmse:.3f} m\n"
        f"Median: {median:.3f} m\n"
        f"NMAD: {nmad:.3f} m"
    ),
    transform=axs[0].transAxes,
    verticalalignment='top',
    bbox=dict(
        facecolor='white',
        alpha=0.8
    )
)

# ------------------------------------------------------------
# HISTOGRAM
# ------------------------------------------------------------

axs[1].hist(
    res,
    bins=150,
    alpha=0.8
)

axs[1].axvline(
    np.median(res),
    color='red',
    linestyle='--',
    linewidth=2,
    label=f"Median = {np.median(res):.3f} m"
)

axs[1].axvline(
    0,
    color='black',
    linestyle='-',
    linewidth=1
)

axs[1].set_xlabel("Residuals (Corrected DEM − Reference DEM) [m]")
axs[1].set_ylabel("Frequency")

axs[1].set_title(
    "Residual Distribution on Stable Ground"
)

axs[1].legend()

# ------------------------------------------------------------
# SAVE
# ------------------------------------------------------------

plt.tight_layout()

plt.savefig(
    f"{output_path}stable_ground_coregistration_analysis.png",
    dpi=300
)

plt.show()

# ============================================================
# FIGURE 2 — BLAND-ALTMAN
# ============================================================

mean_elev = (ref + tar) / 2

fig, ax = plt.subplots(
    figsize=(10, 6)
)

n_ba = min(10000, len(mean_elev))

idx_ba = np.random.choice(
    len(mean_elev),
    n_ba,
    replace=False
)

ax.scatter(
    mean_elev[idx_ba],
    res[idx_ba],
    s=5,
    alpha=0.3
)

ax.axhline(
    np.median(res),
    color='red',
    linewidth=2,
    label=f"Median = {np.median(res):.3f} m"
)

ax.axhline(
    np.median(res) + 1.96 * nmad,
    color='red',
    linestyle=':',
    linewidth=1,
    label=f"±1.96·NMAD"
)

ax.axhline(
    np.median(res) - 1.96 * nmad,
    color='red',
    linestyle=':',
    linewidth=1
)

ax.axhline(
    0,
    color='black',
    linestyle='--',
    linewidth=1
)

ax.set_xlabel("Mean Elevation (m)")
ax.set_ylabel("Residuals (m)")

ax.set_title(
    "Bland-Altman Plot — Stable Ground"
)

ax.legend()

plt.tight_layout()

plt.savefig(
    f"{output_path}bland_altman_stable_ground.png",
    dpi=300
)

plt.show()

print("\nAll plots saved.")
```

```python
import numpy as np
import rasterio
import geopandas as gpd
import matplotlib.pyplot as plt

from rasterio.features import rasterize
from rasterio.warp import reproject, Resampling
from scipy.stats import median_abs_deviation

# ============================================================
# INPUT PATHS
# ============================================================

dem_reference_path = (
    "/isipd/projects/Response/GIS_RS_projects/"
    "Masterarbeit_Till_Weiss/results/"
    "Inuvik_Airport_2023/DSM/"
    "Inuvik_Airport_DSM_2m.tif"
)

dem_target_path = (
    "/isipd/projects/Response/GIS_RS_projects/"
    "Masterarbeit_Till_Weiss/results/"
    "Inuvik_Airport_2025/DSM/"
    "Inuvik_Airport_DSM_2m.tif"
)

stable_ground_gpkg = (
    "/isipd/projects/Response/GIS_RS_projects/"
    "Masterarbeit_Till_Weiss/AOI/StableReferences/"
    "Inuvik_Airport_Stable_Ground.gpkg"
)

# ============================================================
# LOAD REFERENCE DEM
# ============================================================

with rasterio.open(dem_reference_path) as src_ref:

    dem_ref = src_ref.read(1).astype(np.float32)

    ref_transform = src_ref.transform
    ref_crs = src_ref.crs
    ref_shape = src_ref.shape

    nodata_ref = src_ref.nodata

# ============================================================
# LOAD + ALIGN TARGET DEM
# ============================================================

with rasterio.open(dem_target_path) as src_tgt:

    nodata_tgt = src_tgt.nodata

    dem_tgt = np.full(
        ref_shape,
        np.nan,
        dtype=np.float32
    )

    reproject(
        source=rasterio.band(src_tgt, 1),
        destination=dem_tgt,

        src_transform=src_tgt.transform,
        src_crs=src_tgt.crs,

        dst_transform=ref_transform,
        dst_crs=ref_crs,

        src_nodata=nodata_tgt,
        dst_nodata=np.nan,

        resampling=Resampling.bilinear
    )

# ============================================================
# HANDLE NODATA
# ============================================================

if nodata_ref is not None:
    dem_ref[dem_ref == nodata_ref] = np.nan

dem_ref[~np.isfinite(dem_ref)] = np.nan
dem_tgt[~np.isfinite(dem_tgt)] = np.nan

# ============================================================
# COMPUTE dDSM
# ============================================================

ddsm = dem_tgt - dem_ref

# ============================================================
# VALID MASK
# ============================================================

valid_mask = (
    np.isfinite(dem_ref)
    & np.isfinite(dem_tgt)
    & np.isfinite(ddsm)
)

# ============================================================
# LOAD STABLE GROUND
# ============================================================

stable_gdf = gpd.read_file(stable_ground_gpkg)

stable_gdf = stable_gdf.to_crs(ref_crs)

stable_raster = rasterize(
    [(geom, 1) for geom in stable_gdf.geometry],
    out_shape=ref_shape,
    transform=ref_transform,
    fill=0,
    dtype="uint8"
)

stable_mask = (stable_raster == 1) & valid_mask

# ============================================================
# EXTRACT VALUES
# ============================================================

# All pixels
ref_all = dem_ref[valid_mask]
tgt_all = dem_tgt[valid_mask]
ddsm_all = ddsm[valid_mask]

# Stable pixels
ref_stable = dem_ref[stable_mask]
tgt_stable = dem_tgt[stable_mask]
ddsm_stable = ddsm[stable_mask]

# ============================================================
# STATISTICS
# ============================================================

median = np.nanmedian(ddsm_stable)

std = np.nanstd(ddsm_stable)

rmse = np.sqrt(np.nanmean(ddsm_stable**2))

nmad = 1.4826 * median_abs_deviation(
    ddsm_stable,
    scale=1,
    nan_policy="omit"
)

# ============================================================
# PERCENTILE CLIPPING
# ============================================================

p1, p99 = np.percentile(ddsm_all, [1, 99])

# ============================================================
# FIGURE 1
# DEM AGREEMENT DIAGNOSTICS
# ============================================================

fig, axes = plt.subplots(
    2,
    2,
    figsize=(14, 12)
)

# ============================================================
# PANEL 1 — ALL PIXELS AGREEMENT
# ============================================================

ax = axes[0, 0]

hb = ax.hexbin(
    ref_all,
    tgt_all,
    gridsize=180,
    bins="log",
    mincnt=3,
    linewidths=0
)

min_val = min(
    np.nanmin(ref_all),
    np.nanmin(tgt_all)
)

max_val = max(
    np.nanmax(ref_all),
    np.nanmax(tgt_all)
)

ax.plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    linewidth=1.5,
    color="red"
)

ax.set_title("All Pixels")
ax.set_xlabel("2023 DSM Elevation (m)")
ax.set_ylabel("2025 DSM Elevation (m)")

cb = fig.colorbar(hb, ax=ax)
cb.set_label("log10(N pixels)")

# ============================================================
# PANEL 2 — STABLE GROUND AGREEMENT
# ============================================================

ax = axes[0, 1]

hb = ax.hexbin(
    ref_stable,
    tgt_stable,
    gridsize=180,
    bins="log",
    mincnt=3,
    linewidths=0
)

min_val = min(
    np.nanmin(ref_stable),
    np.nanmin(tgt_stable)
)

max_val = max(
    np.nanmax(ref_stable),
    np.nanmax(tgt_stable)
)

ax.plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    linewidth=1.5,
    color="red"
)

ax.set_title("Stable Ground")
ax.set_xlabel("2023 DSM Elevation (m)")
ax.set_ylabel("2025 DSM Elevation (m)")

cb = fig.colorbar(hb, ax=ax)
cb.set_label("log10(N pixels)")

# ============================================================
# PANEL 3 — ALL PIXEL DISTRIBUTION
# ============================================================

ax = axes[1, 0]

ax.hist(
    ddsm_all,
    bins=250,
    density=True,
    alpha=0.8
)

ax.axvline(
    0,
    linestyle="--",
    linewidth=1,
    color="black"
)

ax.set_xlim(p1, p99)

ax.set_title("All Pixel dDSM Distribution")
ax.set_xlabel("Elevation Change (m)")
ax.set_ylabel("Density")

# ============================================================
# PANEL 4 — STABLE GROUND DISTRIBUTION
# ============================================================

ax = axes[1, 1]

ax.hist(
    ddsm_stable,
    bins=250,
    density=True,
    alpha=0.8
)

ax.axvline(
    0,
    linestyle="--",
    linewidth=1,
    color="black"
)

ax.set_xlim(p1, p99)

stats_text = (
    f"Median = {median:.3f} m\n"
    f"NMAD = {nmad:.3f} m\n"
    f"STD = {std:.3f} m\n"
    f"RMSE = {rmse:.3f} m"
)

ax.text(
    0.02,
    0.98,
    stats_text,
    transform=ax.transAxes,
    va="top",
    bbox=dict(
        facecolor="white",
        alpha=0.85
    )
)

ax.set_title("Stable Ground dDSM Distribution")
ax.set_xlabel("Elevation Change (m)")
ax.set_ylabel("Density")

# ============================================================
# FINAL LAYOUT FIGURE 1
# ============================================================

fig.suptitle(
    "DEM Agreement Diagnostics",
    fontsize=16
)

plt.tight_layout()

# ============================================================
# FIGURE 2
# FULL dDSM DISTRIBUTION
# ============================================================

fig2, ax = plt.subplots(
    figsize=(12, 6)
)

ax.hist(
    ddsm_all,
    bins=300,
    density=True,
    alpha=0.4,
    label="All pixels"
)

ax.hist(
    ddsm_stable,
    bins=300,
    density=True,
    alpha=0.7,
    label="Stable ground"
)

ax.axvline(
    0,
    linestyle="--",
    linewidth=1.5,
    color="black"
)

ax.set_xlim(p1, p99)

stats_text = (
    f"Median = {median:.3f} m\n"
    f"NMAD = {nmad:.3f} m\n"
    f"STD = {std:.3f} m\n"
    f"RMSE = {rmse:.3f} m"
)

ax.text(
    0.02,
    0.98,
    stats_text,
    transform=ax.transAxes,
    va="top",
    bbox=dict(
        facecolor="white",
        alpha=0.85
    )
)

ax.set_title("dDSM Change Distribution")
ax.set_xlabel("Elevation Change (m)")
ax.set_ylabel("Density")

ax.legend()

plt.tight_layout()

# ============================================================
# SHOW
# ============================================================

plt.show()
```

```python
import numpy as np
import rasterio
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

from rasterio.features import rasterize
from rasterio.warp import reproject, Resampling
from scipy.stats import median_abs_deviation

# ============================================================
# PUBLICATION STYLE
# ============================================================

plt.rcParams.update({
    "font.family":      "sans-serif",
    "font.size":        10,
    "axes.titlesize":   11,
    "axes.labelsize":   10,
    "xtick.labelsize":  9,
    "ytick.labelsize":  9,
    "axes.spines.top":  False,
    "axes.spines.right": False,
    "axes.linewidth":   0.8,
    "xtick.major.width": 0.8,
    "ytick.major.width": 0.8,
    "figure.dpi":       150,
    "savefig.dpi":      300,
    "savefig.bbox":     "tight",
    "legend.framealpha": 0.9,
    "legend.edgecolor": "0.8",
    "legend.fontsize":  9,
})

BLUE   = "#3A7DC9"
TEAL   = "#1D9E75"
GRAY   = "#888780"
RED    = "#CC3333"

# ============================================================
# INPUT PATHS
# ============================================================

dem_reference_path = (
    "/isipd/projects/Response/GIS_RS_projects/"
    "Masterarbeit_Till_Weiss/results/"
    "Inuvik_Airport_2023/DTM/"
    "Inuvik_Airport_DTM_2m.tif"
)

dem_target_path = (
    "/isipd/projects/Response/GIS_RS_projects/"
    "Masterarbeit_Till_Weiss/results/"
    "Inuvik_Airport_2025/DTM/"
    "Inuvik_Airport_DTM_2m.tif"
)

stable_ground_gpkg = (
    "/isipd/projects/Response/GIS_RS_projects/"
    "Masterarbeit_Till_Weiss/AOI/StableReferences/"
    "Inuvik_Airport_Stable_Ground.gpkg"
)

# ============================================================
# LOAD REFERENCE DEM
# ============================================================

with rasterio.open(dem_reference_path) as src_ref:
    dem_ref     = src_ref.read(1).astype(np.float32)
    ref_transform = src_ref.transform
    ref_crs     = src_ref.crs
    ref_shape   = src_ref.shape
    nodata_ref  = src_ref.nodata

# ============================================================
# LOAD + ALIGN TARGET DEM
# ============================================================

with rasterio.open(dem_target_path) as src_tgt:
    nodata_tgt = src_tgt.nodata
    dem_tgt = np.full(ref_shape, np.nan, dtype=np.float32)
    reproject(
        source=rasterio.band(src_tgt, 1),
        destination=dem_tgt,
        src_transform=src_tgt.transform,
        src_crs=src_tgt.crs,
        dst_transform=ref_transform,
        dst_crs=ref_crs,
        src_nodata=nodata_tgt,
        dst_nodata=np.nan,
        resampling=Resampling.bilinear
    )

# ============================================================
# HANDLE NODATA
# ============================================================

if nodata_ref is not None:
    dem_ref[dem_ref == nodata_ref] = np.nan

dem_ref[~np.isfinite(dem_ref)] = np.nan
dem_tgt[~np.isfinite(dem_tgt)] = np.nan

# ============================================================
# COMPUTE dDSM
# ============================================================

ddsm = dem_tgt - dem_ref

# ============================================================
# MASKS
# ============================================================

valid_mask = (
    np.isfinite(dem_ref)
    & np.isfinite(dem_tgt)
    & np.isfinite(ddsm)
)

stable_gdf    = gpd.read_file(stable_ground_gpkg).to_crs(ref_crs)
stable_raster = rasterize(
    [(geom, 1) for geom in stable_gdf.geometry],
    out_shape=ref_shape,
    transform=ref_transform,
    fill=0,
    dtype="uint8"
)
stable_mask = (stable_raster == 1) & valid_mask

# ============================================================
# EXTRACT VALUES
# ============================================================

ref_all    = dem_ref[valid_mask]
tgt_all    = dem_tgt[valid_mask]
ddsm_all   = ddsm[valid_mask]

ref_stable  = dem_ref[stable_mask]
tgt_stable  = dem_tgt[stable_mask]
ddsm_stable = ddsm[stable_mask]

# ============================================================
# STATISTICS  (stable ground only — these are accuracy metrics)
# ============================================================

med_stable  = np.nanmedian(ddsm_stable)
std_stable  = np.nanstd(ddsm_stable)
rmse_stable = np.sqrt(np.nanmean(ddsm_stable**2))
nmad_stable = 1.4826 * median_abs_deviation(
    ddsm_stable, scale=1, nan_policy="omit"
)
n_all    = int(np.sum(valid_mask))
n_stable = int(np.sum(stable_mask))

# Scene-wide median (headline change number)
med_all = np.nanmedian(ddsm_all)

# ============================================================
# CLIP LIMITS
# ============================================================

p1_all,  p99_all  = np.percentile(ddsm_all,    [1, 99])
p1_stab, p99_stab = np.percentile(ddsm_stable, [1, 99])

# Shared x-axis for histograms in Figure 1  (stable-ground range)
xlim_hist = (
    min(p1_all, p1_stab)   - 0.05,
    max(p99_all, p99_stab) + 0.05,
)

# ============================================================
# HELPER — consistent scatter panel
# ============================================================

def scatter_panel(ax, x, y, title, colorbar=True):
    lim_lo = min(np.nanmin(x), np.nanmin(y))
    lim_hi = max(np.nanmax(x), np.nanmax(y))
    pad    = (lim_hi - lim_lo) * 0.02

    hb = ax.hexbin(
        x, y,
        gridsize=120,
        bins="log",
        mincnt=2,
        linewidths=0,
        cmap="Blues"
    )
    ax.plot(
        [lim_lo, lim_hi],
        [lim_lo, lim_hi],
        linestyle="--", linewidth=1.2, color=RED, label="1:1"
    )
    ax.set_xlim(lim_lo - pad, lim_hi + pad)
    ax.set_ylim(lim_lo - pad, lim_hi + pad)
    ax.set_aspect("equal")
    ax.set_title(title, fontweight="normal")
    ax.set_xlabel("2023 DSM elevation (m)")
    ax.set_ylabel("2025 DSM elevation (m)")

    if colorbar:
        cb = plt.colorbar(hb, ax=ax, pad=0.02, shrink=0.85)
        cb.set_label("log$_{10}$(N pixels)", fontsize=8)
        cb.ax.tick_params(labelsize=8)

    return hb

# ============================================================
# FIGURE 1 — DEM Agreement Diagnostics
# ============================================================

fig1, axes = plt.subplots(
    2, 2,
    figsize=(12, 11),
    gridspec_kw={"hspace": 0.42, "wspace": 0.38}
)

# — (a) All pixels scatter —
scatter_panel(axes[0, 0], ref_all, tgt_all, "(a) All pixels")

# — (b) Stable ground scatter —
scatter_panel(axes[0, 1], ref_stable, tgt_stable, "(b) Stable ground")

# — (c) All pixels distribution —
ax = axes[1, 0]
ax.hist(
    ddsm_all,
    bins=200,
    range=xlim_hist,
    density=True,
    color=BLUE,
    alpha=0.75,
    linewidth=0
)
ax.axvline(0,        color="black", linewidth=1.0, linestyle="--", zorder=3)
ax.axvline(med_all,  color=RED,     linewidth=1.2, linestyle="-",  zorder=3,
           label=f"Scene median = {med_all:.3f} m")
ax.set_xlim(xlim_hist)
ax.set_title("(c) All pixels — dDSM distribution", fontweight="normal")
ax.set_xlabel("Elevation change (m)")
ax.set_ylabel("Density")
ax.text(
    0.98, 0.97,
    f"n = {n_all:,}",
    transform=ax.transAxes,
    ha="right", va="top",
    fontsize=8.5,
    color=GRAY
)
ax.legend(loc="upper left", handlelength=1.4)

# — (d) Stable ground distribution —
ax = axes[1, 1]

# Shaded NMAD bands
for k, alpha in [(1, 0.18), (2, 0.09)]:
    ax.axvspan(
        med_stable - k * nmad_stable,
        med_stable + k * nmad_stable,
        color=TEAL,
        alpha=alpha,
        zorder=1
    )

ax.hist(
    ddsm_stable,
    bins=120,
    range=(p1_stab, p99_stab),
    density=True,
    color=TEAL,
    alpha=0.80,
    linewidth=0,
    zorder=2
)
ax.axvline(0,          color="black",  linewidth=1.0, linestyle="--", zorder=4)
ax.axvline(med_stable, color=RED,      linewidth=1.2, linestyle="-",  zorder=4,
           label=f"Median = {med_stable:.3f} m")

ax.set_xlim(p1_stab - 0.05, p99_stab + 0.05)
ax.set_title("(d) Stable ground — dDSM distribution", fontweight="normal")
ax.set_xlabel("Elevation change (m)")
ax.set_ylabel("Density")

# Stats box
stats_text = (
    f"Median  = {med_stable:+.3f} m\n"
    f"NMAD   = {nmad_stable:.3f} m\n"
    f"STD      = {std_stable:.3f} m\n"
    f"RMSE  = {rmse_stable:.3f} m\n"
    f"n = {n_stable:,}"
)
ax.text(
    0.97, 0.97,
    stats_text,
    transform=ax.transAxes,
    ha="right", va="top",
    fontsize=8.5,
    linespacing=1.55,
    bbox=dict(
        boxstyle="round,pad=0.4",
        facecolor="white",
        edgecolor="0.75",
        alpha=0.90
    )
)

# Custom legend for NMAD bands
legend_elements = [
    Line2D([0], [0], color=RED,   linewidth=1.2, label=f"Median = {med_stable:+.3f} m"),
    Line2D([0], [0], color="black", linewidth=1.0, linestyle="--", label="Zero change"),
    Patch(facecolor=TEAL, alpha=0.35, label="±1 NMAD"),
    Patch(facecolor=TEAL, alpha=0.20, label="±2 NMAD"),
]
ax.legend(handles=legend_elements, loc="upper left", handlelength=1.4)

fig1.suptitle(
    "DEM Agreement Diagnostics",
    fontsize=13,
    fontweight="normal",
    y=1.01
)

fig1.savefig("fig1_dem_agreement.png")
print("Saved fig1_dem_agreement.png")

# ============================================================
# FIGURE 2 — Scene-wide dDSM change distribution
# ============================================================

fig2, ax = plt.subplots(figsize=(10, 5))

# — stable NMAD bands as reference for detectability —
for k, alpha, label in [
    (2, 0.12, "±2 NMAD (stable)"),
    (1, 0.22, "±1 NMAD (stable)")
]:
    ax.axvspan(
        med_stable - k * nmad_stable,
        med_stable + k * nmad_stable,
        color=TEAL,
        alpha=alpha,
        zorder=1,
        label=label
    )

# — All-pixel distribution —
ax.hist(
    ddsm_all,
    bins=250,
    range=(p1_all, p99_all),
    density=True,
    color=BLUE,
    alpha=0.55,
    linewidth=0,
    label=f"All pixels (n = {n_all:,})",
    zorder=2
)

# — Stable ground overlay —
ax.hist(
    ddsm_stable,
    bins=120,
    range=(p1_all, p99_all),
    density=True,
    color=TEAL,
    alpha=0.85,
    linewidth=0,
    label=f"Stable ground (n = {n_stable:,})",
    zorder=3
)

# — Reference lines —
ax.axvline(0,       color="black", linewidth=1.0, linestyle="--", zorder=5,
           label="Zero change")
ax.axvline(med_all, color=RED,     linewidth=1.4, linestyle="-",  zorder=5,
           label=f"Scene median = {med_all:+.3f} m")

ax.set_xlim(p1_all, p99_all)
ax.set_xlabel("Elevation change (m)")
ax.set_ylabel("Density")
ax.set_title(
    "Scene-wide elevation change distribution (dDSM, 2023–2025)",
    fontweight="normal"
)

# Stats box — stable ground metrics
stats_text = (
    f"Stable ground:\n"
    f"  Median  = {med_stable:+.3f} m\n"
    f"  NMAD   = {nmad_stable:.3f} m\n"
    f"  RMSE  = {rmse_stable:.3f} m"
)
ax.text(
    0.98, 0.97,
    stats_text,
    transform=ax.transAxes,
    ha="right", va="top",
    fontsize=8.5,
    linespacing=1.55,
    bbox=dict(
        boxstyle="round,pad=0.4",
        facecolor="white",
        edgecolor="0.75",
        alpha=0.90
    )
)

ax.legend(loc="upper left", handlelength=1.6)

fig2.tight_layout()
fig2.savefig("fig2_ddsm_distribution.png")
print("Saved fig2_ddsm_distribution.png")

plt.show()
```
